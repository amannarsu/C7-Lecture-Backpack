```python
import numpy as np
import time
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
```

```python
distances = [1, 13.1, 26.2, 100]
print(type(distances))
print(distances)

numpy_distances = np.array(distances)
print(type(numpy_distances))
print(numpy_distances)
```

```python
# Convert distances in miles to km
conversion_1 = []
for distance in distances:
    conversion_1.append(distance * 1.60934)
print(conversion_1)

conversion_2 = numpy_distances * 1.60934
print(conversion_2)
```

```python
# Function to perform element-wise multiplication using Python lists
def multiply_list(n):
    arr = list(range(n))
    for i in range(n):
        arr[i] = arr[i] * 1.60934
    return arr

# Function to perform element-wise multiplication using NumPy arrays
def multiply_array(n):
    arr = np.arange(n)
    result = arr * 1.60934
    return result
```

```python
# List of iteration sizes
iterations = [10, 100, 1000, 10000, 100000, 1000000, 10000000, 100000000]

# Lists to store execution times
list_times = []
array_times = []
```

```python
# Perform experiments and measure execution times
for n in iterations:
    # Python list
    start_time = time.time()
    multiply_list(n)
    end_time = time.time()
    list_times.append(end_time - start_time)

    # NumPy array
    start_time = time.time()
    multiply_array(n)
    end_time = time.time()
    array_times.append(end_time - start_time)
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(iterations, list_times, marker='o', label='Python List')
plt.plot(iterations, array_times, marker='o', label='NumPy Array')
plt.xlabel('Number of Elements')
plt.ylabel('Execution Time (seconds)')
plt.title('Performance Comparison: Python List vs NumPy Array')
plt.xscale('log')
plt.yscale('log')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```

```python
# Function to perform matrix multiplication using Python lists
def multiply_matrix_list(n):
    arr1 = [[1] * n for _ in range(n)]
    arr2 = [[2] * n for _ in range(n)]
    result = [[0] * n for _ in range(n)]
    for i in range(n):
        for j in range(n):
            for k in range(n):
                result[i][j] += arr1[i][k] * arr2[k][j]
    return result

# Function to perform matrix multiplication using NumPy arrays
def multiply_matrix_array(n):
    arr1 = np.ones((n, n))
    arr2 = np.full((n, n), 2)
    result = np.matmul(arr1, arr2)
    return result
```

```python
# Perform experiments and measure execution times
matrix_sizes = [10, 100, 200, 300, 400, 500]
list_times = []
array_times = []

for n in matrix_sizes:
    # Python list
    start_time = time.time()
    multiply_matrix_list(n)
    end_time = time.time()
    list_times.append(end_time - start_time)

    # NumPy array
    start_time = time.time()
    multiply_matrix_array(n)
    end_time = time.time()
    array_times.append(end_time - start_time)
```

```python
# Plot the results
plt.figure(figsize=(10, 6))
plt.plot(matrix_sizes, list_times, marker='o', label='Python List')
plt.plot(matrix_sizes, array_times, marker='o', label='NumPy Array')
plt.xlabel('Matrix Size')
plt.ylabel('Execution Time (seconds)')
plt.title('Performance Comparison: Python List vs NumPy Array (Matrix Multiplication)')
plt.grid(True)
plt.legend()
plt.tight_layout()
plt.show()
```
