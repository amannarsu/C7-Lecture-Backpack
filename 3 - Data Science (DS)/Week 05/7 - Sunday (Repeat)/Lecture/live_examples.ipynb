```python
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
sns.get_dataset_names()
```

```python
crashes = sns.load_dataset("car_crashes")
```

```python
crashes.head()
```

```python
crashes.info()
```

```python
crashes.describe()
```

```python
crashes.columns
```

```python
sns.scatterplot(data=crashes, x='alcohol', y='total')
plt.title("Relationship between Alcohol Consumption and Total Accidents")
plt.xlabel("Alcohol Consumption")
plt.ylabel("Total Accidents")
```

```python
# Plot barplot of abbrev compared to total, and make the plot wider
plt.figure(figsize=(20, 6))
sns.barplot(data=crashes, x='abbrev', y='total')
plt.title("Total Accidents by State")
plt.xlabel("State")
plt.ylabel("Total Accidents")
plt.show()
```

```python
plt.figure(figsize=(20, 6))
sns.barplot(data=crashes, x='abbrev', y='speeding')
plt.title("Speeding by State")
plt.xlabel("State")
plt.ylabel("Speeding")
plt.show()
```

```python
crashes.drop(columns=['abbrev'], inplace=True)
```

```python
crashes.head()
```

```python
crashes.iloc[:, :3].head()
```

```python
corr_matrix = crashes.iloc[:, :3].corr()
sns.heatmap(corr_matrix, annot=True, cmap="coolwarm")
plt.title("Correlation Matrix")
plt.show()
```

```python
sns.pairplot(data=crashes.iloc[:, :3])
plt.show()
```

```python
sns.jointplot(data=crashes, x='alcohol', y='speeding', kind='reg')
plt.show()
```

```python
fig, axs = plt.subplots(1,2, figsize=(10,5))
sns.scatterplot(x='alcohol', y='total', data=crashes, ax=axs[0])
sns.scatterplot(x='speeding', y='total', data=crashes, ax=axs[1])
plt.show()
```
